```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

def read_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(dict(row))
    return data

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
menu = get_file(menu_filepath)

# @TODO: Read in the sales data into the sales list
sales = get_file(sales_filepath)

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object
for sales_row in sales:    
    
    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
    quantity = int(sales_row["Quantity"])
    sales_item = sales_row["Menu_Item"]
    
    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
    if sales_item not in list(report.keys()):
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}

    # @TODO: For every row in our sales data, loop over the menu records to determine a match
    # @NOTE: Originally, the assessment logic makes the number of impressions in the console equal to 
    #    the number of items in menu_data multiplied by the number of items in sales_data. What breaks the system.
    #    The variable is_sales_item_found was added as a quick fix to a logic error in the assignment steps. 
    #    It is TRUE if the item of sales_data is found on the menu_data
    is_sales_item_found = False
    for menu_row in menu:
        
        # Item,Category,Description,Price,Cost
        # @TODO: Initialize menu data variables
        item = menu_row["item"]
        price = float(menu_row["price"])
        cost = float(menu_row["cost"])

        # @TODO: Calculate profit of each item in the menu data
        profit = price - cost

        # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item
        if item == sales_item:
            # @TODO: Print out matching menu data
            # print(f"'{sales_item}' from sales is equal the item '{item}' from the Menu")

            # sales_item was found in items on the menu. Changing it to TRUE
            is_sales_item_found = True;
            
            # @TODO: Cumulatively add up the metrics for each item key
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity

        # @TODO: Else, the sales item does not equal any of the item in the menu data, therefore no match
        # else:
        #      print(f"{sales_item} does not equal {item}! NO MATCH!")

    # The sales item does not equal any of the item in the menu data (FIXING THE LOGIC). 
    if not is_sales_item_found:
        print(f"{sales_item} does not found on the Menu! NO MATCH!")
        

    # @TODO: Increment the row counter by 1
    row_count += 1


# @TODO: Print total number of records in sales data
print(f"total number of records in sales data:{row_count}")


# @TODO: Write out report to a text file (won't appear on the command line output)

def export_data_txt(file_path, report):
    """
    Export report to a file.
    """    
    with open(file_path, 'w') as file:
        file.write(str(report))
            
export_data_txt('report.txt', report)
```
